```python
from selenium import webdriver 
from selenium.webdriver.common.by import By
import bs4
import pandas as pd
```

```python
driver_path = '/mnt/c/Users/forcs/Documents/workspace_w/chromedriver.exe'
url = 'https://oryun.org/tv/?menu=57&tv_type=photo&yy=&page=1'

driver = webdriver.Chrome(driver_path)
driver.implicitly_wait(3)

driver.get(url)
```

```python
# 전체 게시판 <section class="media_board_area media_board_100 sol_media_board_area">
# 컨텐츠 영역 <div class="media_board_cont">
# thead
# tbody > tr class="media_board_list_line" > td.text
```

```python
columns = ["제목","본문말씀","설교자","날짜"]

data = []
bottom = driver.find_element_by_class_name("paging")
bottom_spans = driver.find_element_by_class_name("paging").find_elements_by_tag_name("span")
mnum = bottom.find_element_by_class_name("m_num")
# current, last = map(int, bottom.text.split()[1].split("/"))

current = int(mnum.find_element_by_tag_name("input").get_property("value"))
last = int(mnum.get_attribute('innerHTML').split("/")[-1].strip())

# for b in bottom:
print(current, last)
```

```python
data = []
while current < last:
    url = f'https://oryun.org/tv/?menu=57&tv_type=photo&yy=&page={current}'
    driver.get(url)

    board = driver.find_element_by_id("tv_list_html")
    spans = board.find_elements_by_tag_name("span")
    # 2, 8, 14, 20
    tmp_data =[]
    for idx,span in enumerate(spans):
        if (idx-2) % 6 == 0:
            tmp_data.append(span.text.strip())
        elif (idx-2) % 6 == 1:
            sp = span.text.split("|")
            tmp_data.append(sp[0].strip())
            tmp_data.append(sp[1].strip())
        elif (idx-2) % 6 == 2:
            tmp_data.append(span.text.strip())
            data.append(tmp_data)
            tmp_data = []
    current += 1
driver.close()
```
