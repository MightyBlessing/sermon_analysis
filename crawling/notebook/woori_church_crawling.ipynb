```python
from selenium import webdriver 
from selenium.webdriver.common.by import By
import bs4
import pandas as pd
```

```python
driver_path = '/mnt/c/Users/forcs/Documents/workspace_w/chromedriver.exe'
url = 'http://www.woorichurch.org/wch/media_board/list.asp?board_idx=101'

driver = webdriver.Chrome(driver_path)
driver.implicitly_wait(3)

driver.get(url)
```

```python
# 전체 게시판 <section class="media_board_area media_board_100 sol_media_board_area">
# 컨텐츠 영역 <div class="media_board_cont">
# thead
# tbody > tr class="media_board_list_line" > td.text
```

```python
board = driver.find_element_by_class_name("media_board_list_tbl")
ths = board.find_elements_by_tag_name("th")

for th in ths:
    print(th.text)
    
columns = [th.text for th in ths]

data = []
bottom = driver.find_element_by_class_name("media_board_bottom")
current, last = map(int, bottom.text.split()[1].split("/"))

while True:    
    print(current,'/',last)
    column_data = []
    
    board = driver.find_element_by_class_name("media_board_list_tbl")
    tds = board.find_elements_by_tag_name("td")

    for idx, td in enumerate(tds):
        if idx % len(columns) == 0:
            if len(column_data) == 6:
                data.append(column_data)
                column_data = []
        column_data.append(td.text)
    
    if  current < last:
        bottom = driver.find_element_by_class_name("media_board_bottom")
        buttons = bottom.find_elements_by_tag_name("a")
        next_button = buttons[1]

        next_button.click()
        driver.implicitly_wait(3)
        
        current +=1
    else:
        break
driver.close()
```
